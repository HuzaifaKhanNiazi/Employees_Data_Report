### 2.2 Duplications Analysis

### 2.3 Outliers Identification

```python
df['Last % Hike Numeric'] = df['Last % Hike'].astype(str).str.rstrip('%').astype(float)
def find_outliers(column_name):

    if column_name not in df.columns:
        print(f"Error: Column '{column_name}' not found.")
        return

    Q1 = df[column_name].quantile(0.25)
    Q3 = df[column_name].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[column_name] < lower_bound) | (df[column_name] > upper_bound)]

    print(f"--- {column_name} ---")
    print(f"Normal Range: {lower_bound} to {upper_bound}")
    print(f"Total Outliers: {len(outliers)}")
    return outliers
```

```python
salary_outliers = find_outliers('Salary')
age_outliers = find_outliers('Age in Yrs.')


hike_outliers = find_outliers('Last % Hike Numeric')
```

### 2.4 Logical Data Errors Check

```python
invalid_joining = df[df['Date of Joining'] < df['Date of Birth']]
print(f"Joining Date before Birth Date errors: {len(invalid_joining)}")

invalid_tenure = df[df['Age in Company (Years)'] > df['Age in Yrs.']]
print(f"Tenure > Age errors: {len(invalid_tenure)}")
```

# 3. Dataset Cleaning

### 3.1 Logical Data Errors Fix

```python
df['Date of Birth'] = pd.to_datetime(df['Date of Birth'], errors='coerce')
df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], errors='coerce')
```

### 3.2 Handling Duplication by Unique Identification

```python
df['Unique_ID'] = df.index + 1

print(df[['Unique_ID', 'Emp ID', 'First Name']].head())
```

# 4. Feature Engineering

### 4.1 Feature Reduction & Extraction

```python
columns_to_drop = ['Emp ID', 'Name Prefix', 'First Name', 'Middle Initial', 'Last Name',
                   'E Mail', "Father's Name", "Mother's Name", "Mother's Maiden Name",
                   'SSN', 'Phone No.', 'User Name', 'Password']


df.drop(columns=[col for col in columns_to_drop if col in df.columns], inplace=True)


df['Date of Joining'] = pd.to_datetime(df['Date of Joining'], errors='coerce')


df['Joining Year'] = df['Date of Joining'].dt.year
df['Joining Month'] = df['Date of Joining'].dt.month_name()
df['Joining Quarter'] = 'Q' + df['Date of Joining'].dt.quarter.astype(str)


if 'Last % Hike' in df.columns and 'Last % Hike Numeric' not in df.columns:
    df['Last % Hike Numeric'] = df['Last % Hike'].astype(str).str.rstrip('%').astype(float)

print("Feature Engineering Completed.")
print("Current Columns:", df.columns.tolist())
```

# 5. Exploratory Data Analysis

### 5.1 Salary Distribution Analysis

```python
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['Salary'])
plt.title('Salary Distribution (Box Plot)')
plt.xlabel('Salary')
plt.show()

print("Salary Statistics:")
print(df['Salary'].describe())
print(f"Skewness: {df['Salary'].skew()}")
```

### 5.2 Gender Diversity & Pay Gap Analysis

```python
plt.figure(figsize=(8, 8))
gender_counts = df['Gender'].value_counts()
plt.pie(gender_counts, labels=gender_counts.index, autopct='%1.1f%%', startangle=140, colors=sns.color_palette('pastel'))
plt.title('Gender Distribution')
plt.show()


plt.figure(figsize=(8, 5))
sns.boxplot(x='Gender', y='Salary', data=df)
plt.title('Salary by Gender (Pay Gap Analysis)')
plt.show()

print("Average Salary by Gender:")
print(df.groupby('Gender')['Salary'].mean())
```
